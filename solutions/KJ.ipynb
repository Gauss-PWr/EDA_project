```python
df_copy = df.copy()
df_to_hist = df_copy.drop(columns=["genre", "artist", "song", "explicit", "mode", "instrumentalness"])
```

```python
sns.set(rc={'figure.figsize':(10,5)})
for column in df_to_hist.keys():
    sns.histplot(data=df, x=column, kde=True)
    plt.tight_layout()
    plt.show()
```

# Boxplot

```python
df_to_hist.plot(kind='box', subplots=True, figsize=(20, 4), color=dict(boxes="Blue"), patch_artist=True)
plt.tight_layout()
plt.show()
```

```python
sns.set(rc={'figure.figsize':(5,5), 'font.size':25})

for column in df_to_hist.keys():
    sns.boxplot(data=df, y = column)
    plt.tight_layout()
    plt.show()
```

```python
sns.set(rc={'figure.figsize':(5,5), 'font.size':25})
sns.boxplot(data=df, y = "popularity", x="explicit")
plt.tight_layout()
plt.show()
```
